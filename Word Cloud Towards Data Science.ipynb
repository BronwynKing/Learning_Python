```python
# Import packages

import wikipedia
import re

# Specify the title of the Wikipedia page
wiki = wikipedia.page('South Bastion Mountain')

# Extract the plain text content of the page
text = wiki.content

# Clean text
text = re.sub(r'==.*?==+', '', text)
text = text.replace('\n', '')
```

```python
# Import package
import matplotlib.pyplot as plt

# Define a function to plot word cloud
def plot_cloud(mydata):
    # Set figure size
    plt.figure(figsize=(40, 30))
    
    # Display image
    plt.imshow(mydata) 
    
    # No axis details
    plt.axis("off");
```

```python
# Import package
from wordcloud import WordCloud, STOPWORDS
# Generate word cloud
dog = WordCloud(width = 3000, height = 2000, random_state=1,
                      background_color='orange', colormap='Pastel1', 
                      collocations=False, stopwords = STOPWORDS).generate(text)
# Plot
plot_cloud(mydata=dog)
```
